```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

#lets pull in a dataset

```python
main_df=pd.read_csv("https://breathecode.herokuapp.com/asset/internal-link?id=418&path=demographic_health_data.csv")
main_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df,test_df=train_test_split(main_df)
```

```python
#dropping columns to reduce data leakage sources
x_train = train_df.drop(columns=['fips','COUNTY_NAME', 'anycondition_prevalence',
 'anycondition_Lower 95% CI','anycondition_Upper 95% CI','anycondition_number',
 'Obesity_prevalence','Obesity_Lower 95% CI','Obesity_Upper 95% CI','Obesity_number',
 'Heart disease_prevalence','Heart disease_Lower 95% CI','Heart disease_Upper 95% CI','Heart disease_number',
 'COPD_prevalence','COPD_Lower 95% CI','COPD_Upper 95% CI','COPD_number',
 'diabetes_prevalence','diabetes_Lower 95% CI','diabetes_Upper 95% CI','diabetes_number',
 'CKD_prevalence','CKD_Lower 95% CI','CKD_Upper 95% CI','CKD_number'])
y_train = train_df['diabetes_prevalence']
```

```python
x_train.isna().sum().sum()
```

```python
x_train_dummies = pd.get_dummies(x_train,drop_first=True)
```

 Let's build a standard Linear Regression algorithm to predict `diabetes_prevalence`

```python
from sklearn.linear_model import LinearRegression
```

```python
#instantiate a linear regression model
#Linear Regression modelini oluturuyoruz, artk onu kullanabiliriz
linreq_model = LinearRegression()
```

```python
#fit the model to our taining data
linreq_model.fit(x_train_dummies,y_train)
```

```python
# make our first set of predicition ( again on the training data)
train_preds = linreq_model.predict(x_train_dummies)
```

```python
from sklearn.metrics import root_mean_squared_error,mean_absolute_error,r2_score
```

RMSE (Root MSE):MSEnin karekk  hata birimi orijinal birimle olur/Hata birimini anlamak istediinde/Ev fiyat tahmininde $ olarak hata lmek istiyorsan

r (Coefficient of Determination):Modelin verideki deiimi ne kadar akladn gsterir/Modelin baarsn genel olarak grmek istediinde
R = 0.9  model hedefi %90 orannda aklyor

MAE (Mean Absolute Error):	Tahminler ile gerek deer arasndaki ortalama mutlak fark/	Genel hatay grmek istediinde/	Model ortalama olarak 10.000$ sapyor mu?

MSE (Mean Squared Error):    Farklar kareye alr  byk hatalar daha ok cezalandrr/   Byk hatalarn ciddi problem olduu durumlarda

rnek

ok pahal evler varsa ve byk hatalar vurgulamak istiyorsan          

```python
print(f"root mean squard error (RMSE):{root_mean_squared_error(y_true=y_train,y_pred=train_preds)}")
print(f"MEAN absolute error (MAE):{mean_absolute_error(y_true=y_train,y_pred=train_preds)}")
print(f"r^2:{r2_score(y_true=y_train,y_pred=train_preds)}")
```

```python
sns.scatterplot(x=y_train,y=train_preds)
sns.scatterplot(x=np.linspace(0,22.5),y=np.linspace(0,22.5))
plt.grid()
```

```python
train_residual= train_preds - y_train
```

```python
sns.scatterplot(x= train_residual.index, y = train_residual)
plt.grid()
plt.xlabel("train_residual_index")
plt.ylabel("train_residual")
```
