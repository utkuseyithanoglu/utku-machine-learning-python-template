```python
from sklearn.model_selection import train_test_split
```

```python
train_df,test_df = train_test_split(main_df)
```

```python
X_train = train_df.drop(columns=['fips','COUNTY_NAME', 'anycondition_prevalence',
 'anycondition_Lower 95% CI','anycondition_Upper 95% CI','anycondition_number',
 'Obesity_prevalence','Obesity_Lower 95% CI','Obesity_Upper 95% CI','Obesity_number',
 'Heart disease_prevalence','Heart disease_Lower 95% CI','Heart disease_Upper 95% CI','Heart disease_number',
 'COPD_prevalence','COPD_Lower 95% CI','COPD_Upper 95% CI','COPD_number',
 'diabetes_prevalence','diabetes_Lower 95% CI','diabetes_Upper 95% CI','diabetes_number',
 'CKD_prevalence','CKD_Lower 95% CI','CKD_Upper 95% CI','CKD_number'])
y_train = train_df['diabetes_prevalence']
```

```python
#create dummy variables for all of our categorcal featruws (in this case the state name)
x_train_dummies=pd.get_dummies(data=X_train,drop_first=True)
```

# lest build a standart regression algorithm to predict diabetes_prevelence

```python
from sklearn.linear_model import LinearRegression
```

```python
#instantiate a linear regression model
linreq_model = LinearRegression()
#fit the model our trainig data
linreq_model.fit(x_train_dummies,y_train)
#make our first set of prediction
train_preds = linreq_model.predict(x_train_dummies)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error ,r2_score
```

```python
print("ON the trainig dataset :")
print(f"mean absolute error :{mean_absolute_error(y_true=y_train,y_pred=train_preds)}")
print(f" root mean squared error :{mean_squared_error(y_true=y_train,y_pred=train_preds)}")
print(f"r2_score :{mean_absolute_error(y_true=y_train,y_pred=train_preds)}")
print(f"mean_squared_error  :{mean_squared_error(y_true=y_train,y_pred=train_preds)}")

```
